```python
csv_path = '/Users/tobi/Library/CloudStorage/OneDrive-Persönlich/Dokumente/Uni/Masterarbeit - Thema/WRDS Export.csv'

df_raw = pd.read_csv(csv_path, sep=';', encoding='utf-8', low_memory=False)

print(f"Shape: {df_raw.shape}")
print(f"Spalten: {len(df_raw.columns)}")
print(f"Zeilen: {len(df_raw)}")
```

```python
# Fehlende Werte pro Spalte (Top 20)
missing = df_raw.isna().sum().sort_values(ascending=False).head(20)
missing_pct = (missing / len(df_raw) * 100).round(1)

missing_df = pd.DataFrame({
    'Fehlende Werte': missing,
    'Prozent': missing_pct
})

print("Top 20 Spalten mit meisten fehlenden Werten:")
missing_df
```

```python
# Wichtige Finanzkennzahlen-Spalten
key_columns = ['conm', 'datadate', 'at', 'revt', 'ebit', 'ebitda', 'ib', 'seq', 'lt', 'emp']
available = [col for col in key_columns if col in df_raw.columns]

df_key = df_raw[available].copy()

print("Verfügbare wichtige Spalten:")
print(available)
print("\nStatistiken:")
df_key.describe()
```
